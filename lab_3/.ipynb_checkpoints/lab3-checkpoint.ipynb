```julia
#tablice wielowymiarowe w Julii
Asmall=[[1.0 0.0]; [0.0 1.0]]
Bsmall=Asmall
#size(Bsmall,1)
```

```julia
# mnożenie macierzy - wersja naiwna
function naive_multiplication(A,B)
C=zeros(Float64,size(A,1),size(B,2))
  for i=1:size(A,1)
    for j=1:size(B,2)
        for k=1:size(A,2)
            C[i,j]=C[i,j]+A[i,k]*B[k,j]
        end
    end
end
C
end
```

```julia
naive_multiplication(Asmall,Bsmall)
```

```julia
A=rand(1000,1000);
B=rand(1000,1000);
```

```julia
# Należy pamiętać o "column-major" dostępie do tablic - 
# pierwszy indeks zmienia się szybciej
# tak jak Matlab, R, Fortran 
# inaczej niz C, Python
A1 = [[1 2]; [3 4]]
vec(A1)
```

```julia
# poprawiona funkcja korzytająca z powyższego oraz z faktu, że
#można zmieniać kolejność operacji dodawania (a co za tym idzie kolejnosc petli).
function better_multiplication( A,B )
C=zeros(Float64,size(A,1),size(B,2))
  for j=1:size(A,2)
    for k=1:size(B,2)
        for i=1:size(A,1)
            C[i,j]=C[i,j]+A[i,k]*B[k,j]
        end
    end
end
C
end
```

```julia
better_multiplication(Asmall, Bsmall)
```

```julia
function test_methods(matrix_size)
    A=rand(matrix_size, matrix_size);
    B=rand(matrix_size, matrix_size);
    naive = @elapsed naive_multiplication(A, B)
    better = @elapsed better_multiplication(A, B)
    blas = @elapsed A*B
    return  naive, better, blas
end
```

```julia
test_methods(1000)
```

```julia
using DataFrames
function test_suite(range)
    results = DataFrame(Size = Int64[], Time = Float64[], Method = String[])
    for i in range
        for j in 1:10
            n, b, bl = test_methods(i)
            push!(results, [i, n, "Naive"])
            push!(results, [i, b, "Better"])
            push!(results, [i, bl, "Blas"])
        end
    end
    results
end
```

```julia
using Statistics
data = test_suite(100:100:1000)
data = by(data, [:Size, :Method],  df->DataFrame(
        Mean=mean(df[:Time]), Deviation=std(df[:Time])))
```

```julia
julia = scatter(data[:Size], data[:Mean], group = data[:Method], yerr = data[:Deviation],
    colour = [:red :blue :green], label = data[:Method], 
    xlabel = "Size", ylabel = "Time [s]", title = "Comparison of matrix multiplication methods")
```
