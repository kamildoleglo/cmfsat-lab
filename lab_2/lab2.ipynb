```julia
using CSV
data = CSV.read("results.csv")
data = by(data, [:LEVEL, :SIZE],  df->DataFrame(MEAN=mean(df[:TIME]), DEVIATION=std(df[:TIME])))
```

```julia
using Plots
data1 = data[data[:LEVEL].== 1, :]
data2 = data[data[:LEVEL].== 2, :]
p1 = scatter(data1[:SIZE], data1[:MEAN], yerr = data1[:DEVIATION], title = "Level 1", color = [:red])
p2 = scatter(data2[:SIZE], data2[:MEAN], yerr = data2[:DEVIATION], title = "Level 2")
plot(p1, p2, layout = (2,1), xlabel = "Size", ylabel = "Time [s]", legend = false, plot_title = "Comparison of level 1 and level 2 BLAS")
# plot_title is not currently implemented, let's wait...
```
